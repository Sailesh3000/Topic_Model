```python
taja_text = soup.find_all("div", attrs={"class":"webfeat"})
listTag = taja_text[0].find("ul")
listTagAll = listTag.find_all("li")
links = []

telugu_uni = soup.find_all("aside", attrs={"class":"thumb-content"})
telugu_uni_links = []
for i in range(len(telugu_uni)):
  telugu_uni_links.append(telugu_uni[i].find("a")["href"])

for i in range(len(telugu_uni_links)):
  links.append(telugu_uni_links[i])
links
```

```python
import csv
from google.colab import drive

drive.mount('/content/drive')

def scrape_page(url):
    try:
        response = requests.get(url)
        response.raise_for_status()

        soup = BeautifulSoup(response.text, 'html.parser')

        p_tag = soup.find('p', class_='srtdes')

        if p_tag:
            return p_tag.get_text(strip=True)
        else:
            return None
    except requests.exceptions.RequestException as e:
        print(f"Error fetching {url}: {e}")
        return None

file_path = 'drive/MyDrive/Colab/scraped_data.csv'
with open(file_path, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)

    writer.writerow(['Content'])

    for url in links:
        data = scrape_page(url)
        if data:
            writer.writerow([data])
        else:
            writer.writerow(['No data found'])

print("Data has been successfully written to 'scraped_data.csv'")
```
